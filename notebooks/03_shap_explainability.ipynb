### IMPORT AND SETUP

```python
import shap
import joblib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import lightgbm as lgb

# Enable JS for SHAP visualizations
shap.initjs()
```

### Load trained model and validation data

```python
model = joblib.load("../models/final_model_lgbm.pkl")
X_val, y_val = joblib.load("../output/val_data.joblib")

print("Model and validation data loaded successfully.")
print("Validation shape:", X_val.shape)
```

### SHAP

```python
# Convert to DataFrame using model's feature names
X_val_df = pd.DataFrame(X_val, columns=model.feature_name_)

# Create SHAP explainer
explainer = shap.TreeExplainer(model)

# Compute SHAP values for the entire validation set
shap_values = explainer.shap_values(X_val_df)

# SHAP Summary Plot for class 1 (churn)
plt.title("SHAP Summary Plot - Global Feature Importance")
shap.summary_plot(shap_values, X_val_df)
```
